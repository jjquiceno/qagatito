```python
import matplotlib.pyplot as plt

# Definicin de datos para cada sprint (mximo de 6 sprints)
sprints = ["Sprint 1", "Sprint 2", "Sprint 3", "Sprint 4", "Sprint 5", "Sprint 6"]

# Datos simulados
tasa_exito_envio = [70, 75, 80, 85, 90, 95]  # En porcentaje
errores_validacion = [15, 12, 10, 8, 5, 3]  # Nmero de errores
verificacion_layout = [60, 70, 80, 85, 90, 95]  # En porcentaje
verificacion_funcionalidad = [65, 75, 80, 85, 90, 95]  # En porcentaje
verificacion_compatibilidad = [50, 65, 75, 80, 90, 95]  # En porcentaje
documentacion_resultados = [40, 60, 70, 80, 90, 100]  # En porcentaje
satisfaccion_usuario = [5, 6, 7, 8, 9, 10]  # Calificacin del 1 al 10

# Configuracin del grfico
plt.figure(figsize=(14, 8))

# Grfico de lneas para cada mtrica
plt.plot(sprints, tasa_exito_envio, marker='o', label='Tasa de xito de envo')
plt.plot(sprints, errores_validacion, marker='o', label='Nmero de errores de validacin')
plt.plot(sprints, verificacion_layout, marker='o', label='Verificacin de Layout y Diseo')
plt.plot(sprints, verificacion_funcionalidad, marker='o', label='Verificacin de Funcionalidad')
plt.plot(sprints, verificacion_compatibilidad, marker='o', label='Verificacin de Compatibilidad')
plt.plot(sprints, documentacion_resultados, marker='o', label='Documentacin de Resultados')
plt.plot(sprints, satisfaccion_usuario, marker='o', label='Satisfaccin del usuario')

# Detalles del grfico
plt.title('Simulacin de Proyecto: Evolucin de Mtricas en Sprints')
plt.xlabel('Sprints')
plt.ylabel('Valores')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
